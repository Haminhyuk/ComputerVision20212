## 1.4.7 Canny Edge Detection

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('cat.jpg',0)
edges = cv2.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread("cat.jpg",0)
blur = cv2.GaussianBlur(img,(5,5),0)
edges1 = cv2.Canny(img,30,255)
edges2 = cv2.Canny(blur,100,250) 

plt.subplot(221),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(blur,cmap = 'gray')
plt.title('Gaussian blur'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(edges2,cmap = 'gray')
plt.title('blur+Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
