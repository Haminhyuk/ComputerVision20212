```python
#load an color image
img = cv2.imread('cat.jpg', 0)
cv2.imshow('cat', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('cat', img)
k = cv2.waitKey(0)
if k == 27:
    cv2.destroyAllWindows()
elif k == ord('s'):
    cv2.imwrite('cat.jpg', img)
    cv2.destroyAllWindows()
```
